```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy import stats
import numpy as np
```

# Mais sobre o projeto:
https://www.freecodecamp.org/learn/data-analysis-with-python/data-analysis-with-python-projects/sea-level-predictor

```python
# Lendo dados
df = pd.read_csv('epa-sea-level.csv')
```

```python
# Fazendo a regressão linear considerando todos os anos disponíveis
res = stats.linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
```

```python
# Fazendo a regressão linear considerando do ano 2000 pra frente
df2 = df[df['Year']>=2000]
res2 = stats.linregress(df2['Year'],df2['CSIRO Adjusted Sea Level'])
```

```python
# Fazendo todos os plots
fig,axs = plt.subplots(figsize=(10,10))

plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')

plt.scatter(data=df,x='Year',y='CSIRO Adjusted Sea Level')
x1 = np.linspace(1880,2050,100)
y1 = res.intercept + res.slope*x1
plt.plot(x1,y1)

x2 = np.linspace(1950,2050,100)
y2 = res2.intercept + res2.slope*x2
plt.plot(x2,y2)
```
